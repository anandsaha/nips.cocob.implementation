This notebook plots the graphs from the pickle file `result_4.pkl`, which has results of running the `word_language_model` program on various optimizers. Check the readme at `word_language_model/Readme.md` to know how `result_4.pkl` was generated.

```python
import pickle
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def read_data(fname):
    with open(fname, "rb") as fp:
        return pickle.load(fp)
```

```python
data = read_data('result_4.pkl')
```

```python
colors = dict()
colors['cocob'] = 'red'
colors['adam'] = 'black'
colors['adadelta'] = 'blue'
colors['rmsprop'] = 'magenta'
colors['adagrad'] = 'green'
```

```python
f = plt.figure(figsize=(10, 7))
for opt in data:
    plt.plot(data[opt][0], label=opt, c=colors[opt])
plt.legend()
plt.suptitle('Word Prediction on PTB - Training Cost', fontsize=16)
plt.ylabel('Perplexity')
plt.xlabel('Epochs')
plt.show()
```

```python
f = plt.figure(figsize=(10, 7))
for opt in data:
    plt.plot(data[opt][1], label=opt, c=colors[opt])
plt.legend()
plt.suptitle('Word Prediction on PTB - Validation Cost', fontsize=16)
plt.ylabel('Perplexity')
plt.xlabel('Epochs')
plt.show()
```
